# Varying the B field

Now we will analyse how the simulation results, looking at diffusion constant D and possible exponent, will change with the B field, keeping other variables fixed.

The previously used functions can now be found in the separate python module 'Diffulyzer'.

```python
%matplotlib inline

from Diffulyzer import *

num_candidates = 300
EMin = 1 * EeV
EMax = 1 * EeV
num_samples = 200
max_trajectory_length = 100 * Mpc
grid_size = 256
grid_spacing = 30 * kpc
lMin = 100 * kpc
lMax = 1000 * kpc
alpha = -11./3.

Brms_min = 10 * nG
Brms_max = 100 * nG
variation_steps = 10
variation_repeat = 3
```

With the above settings we run a bunch of simulations. Now we will skip the trajectories, as they don't tell us anything new. We will keep the <r²> plots to be able to check soundness.

```python
delta_B = (Brms_max - Brms_min) / (variation_steps-1)
Bs = []
r_L = []
Ds = []
exponents = []
for i in range(variation_steps):
    Brms = Brms_min + i * delta_B
    for j in range(variation_repeat):
        nr = 'Fit ' + str(i*variation_repeat+j+1) + '/' + str(variation_steps*variation_repeat)
        plot_title = nr + ', Brms=' + str(Brms / nG) + ' nG, E=' + str(EMin / EeV) + ' EeV'
        Bs.append(Brms / nG)
        r_L.append(larmor_radius(EMin, Brms) / kpc)
        data = sim(num_candidates, num_samples, max_trajectory_length, EMin, EMax, grid_size, grid_spacing, Brms, lMin, lMax, alpha)
        D, exponent = plot_rms(data, plot_title)
        Ds.append(D)
        exponents.append(exponent)
```

An overview of all results shows the dependency on the B field strength.

```python
figure(figsize=(9, 5))
title('D ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(Bs, Ds, s=50)
grid()
ylabel('D')
xlabel('Brms [nG]')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(Bs, exponents, s=50)
grid()
ylabel('exponent')
xlabel('Brms [nG]')
legend()
show()       
```

This looks like a rough dependency of D ~ 1 / B. The exponent stays close to 1 which is the diffusion realm. If it was going up to 2 we were back in the ballistic motion, but here diffusion is a good description of the motion.

For better analysis we will express our B and E setting by the Larmor radius r_L = E / e c Brms.

```python
figure(figsize=(9, 5))
title('D ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(r_L, Ds, s=50)
grid()
ylabel('D')
xlabel('r_L [kpc]')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(r_L, exponents, s=50)
grid()
ylabel('exponent')
xlabel('r_L [kpc]')
legend()
show()  
```

As the Larmor radius is proportional to 1/B we could guess a linear fit for D. But of course we set our spacing for B and not for r_L on the x axis.
