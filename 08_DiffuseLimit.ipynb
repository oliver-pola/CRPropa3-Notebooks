# Limits of diffuse

Now we want to see what happens if we leave the realm of diffusion und go to ballistic motion. This should happen if we further reduce the B field strength so that at last the particles are almost not affected.

The previously used functions can now be found in the separate python module 'Diffulyzer'.

```python
%matplotlib inline

from Diffulyzer import *

num_candidates = 300
EMin = 1 * EeV
EMax = 1 * EeV
num_samples = 200
max_trajectory_length = 3 * Mpc
grid_size = 256
grid_spacing = 30 * kpc
lMin = 100 * kpc
lMax = 1000 * kpc
alpha = -11./3.

Brms_min = 2 * nG
Brms_max = 10 * nG
variation_steps = 9
variation_repeat = 3
```

```python
delta_B = (Brms_max - Brms_min) / (variation_steps-1)
Bs = []
r_L = []
Ds = []
exponents = []
datas = []
Brms = Brms_max * 2;
for i in range(variation_steps):
    Brms = Brms_min + i * delta_B
    for j in range(variation_repeat):
        nr = 'Fit ' + str(i*variation_repeat+j+1) + '/' + str(variation_steps*variation_repeat)
        plot_title = nr + ', Brms=' + str(Brms / nG) + ' nG, E=' + str(EMin / EeV) + ' EeV'
        Bs.append(Brms / nG)
        r_L.append(larmor_radius(EMin, Brms) / kpc)
        data = sim(num_candidates, num_samples, max_trajectory_length, EMin, EMax, grid_size, grid_spacing, Brms, lMin, lMax, alpha)
        datas.append(data)
        D, exponent = plot_rms(data, plot_title)
        Ds.append(D)
        exponents.append(exponent)
```

An overview of all results shows the dependency on the B field strength.

```python
figure(figsize=(9, 5))
title('D ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(Bs, Ds, s=50)
grid()
ylabel('D')
xlabel('Brms [nG]')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(Bs, exponents, s=50)
grid()
ylabel('exponent')
xlabel('Brms [nG]')
legend()
show()    

figure(figsize=(9, 5))
title('D ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(r_L, Ds, s=50)
grid()
ylabel('D')
xlabel('r_L [kpc]')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ Brms, E=' + str(EMin / EeV) + ' EeV')
scatter(r_L, exponents, s=50)
grid()
ylabel('exponent')
xlabel('r_L [kpc]')
legend()
show()
```

We can see the exponent going back to almost 2 again for the lower B values.
