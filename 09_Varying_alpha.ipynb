```python
from Diffulyzer import *

num_candidates = 300
EMin = 1 * EeV
EMax = 1 * EeV
num_samples = 200
max_trajectory_length = 100 * Mpc
grid_size = 256
grid_spacing = 30 * kpc
lMin = 100 * kpc
lMax = 1000 * kpc
Brms = 100 * nG

#alpha = -11./3.
alpha_min = -8.0
alpha_max = 3.0
variation_steps = 13
variation_repeat = 3
```

```python
delta_a = (alpha_max - alpha_min) / (variation_steps-1)
Alphas = []
Ds = []
exponents = []
for i in range(variation_steps):
    alpha = alpha_min + i * delta_a
    for j in range(variation_repeat):
        nr = 'Fit ' + str(i*variation_repeat+j+1) + '/' + str(variation_steps*variation_repeat)
        plot_title = nr + ', Brms=' + str(Brms / nG) + ' nG, E=' + str(EMin / EeV) + ' EeV' + ' alpha = ' + str(alpha)
        Alphas.append(alpha)
        data = sim(num_candidates, num_samples, max_trajectory_length, EMin, EMax, grid_size, grid_spacing, Brms, lMin, lMax, alpha)
        D, exponent = plot_rms(data, plot_title)
        Ds.append(D)
        exponents.append(exponent)
```

```python
figure(figsize=(9, 5))
title('D ~ alpha, E=' + str(EMin / EeV) + ' EeV' + 'Brms =' + str(Brms / nG) + ' nG')
scatter(Alphas, Ds, s=50)
grid()
ylabel('D')
xlabel('Alpha')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ alpha, E=' + str(EMin / EeV) + ' EeV'+ 'Brms =' + str(Brms / nG) + ' nG')
scatter(Alphas, exponents, s=50)
grid()
ylabel('exponent')
xlabel('alpha')
legend()
show()   
```
