# Varying the particles energy

As we previously looked at the resulting diffusion constant by varying the B field strength, we will now keep that fixed and vary the energy each particle has. But we will still keep all particles in one simulation run to have the same energy.

The functions can be found in the separate python module 'Diffulyzer'.

```python
%matplotlib inline

from Diffulyzer import *

num_candidates = 300
num_samples = 200
max_trajectory_length = 100 * Mpc
grid_size = 256
grid_spacing = 30 * kpc
Brms = 100 * nG
lMin = 100 * kpc
lMax = 1000 * kpc
alpha = -11./3.

E_min = 1 * EeV
E_max = 10 * EeV
variation_steps = 10
variation_repeat = 3
```

With the above settings we run a bunch of simulations. Now we will skip the trajectories, as they don't tell us anything new. We will keep the <r²> plots to be able to check soundness.

```python
delta_E = (E_max - E_min) / (variation_steps-1)
Es = []
Ds = []
exponents = []
for i in range(variation_steps):
    E = E_min + i * delta_E
    for j in range(variation_repeat):
        nr = 'Fit ' + str(i*variation_repeat+j+1) + '/' + str(variation_steps*variation_repeat)
        plot_title = nr + ', Brms=' + str(Brms / nG) + ' nG, E=' + str(E / EeV) + ' EeV'
        Es.append(E / EeV)
        data = sim(num_candidates, num_samples, max_trajectory_length, E, E, grid_size, grid_spacing, Brms, lMin, lMax, alpha)
        D, exponent = plot_rms(data, plot_title)
        Ds.append(D)
        exponents.append(exponent)
```

An overview of all results shows the dependency on the B field strength.

```python
figure(figsize=(9, 5))
title('D ~ E, Brms=' + str(Brms / nG) + ' nG')
scatter(Es, Ds, s=50)
grid()
ylabel('D')
xlabel('E [EeV]')
legend()
show()

figure(figsize=(9, 5))
title('exponent ~ E, Brms=' + str(Brms / nG) + ' nG')
scatter(Es, exponents, s=50)
grid()
ylabel('exponent')
xlabel('E [EeV]')
legend()
show()    
```

This looks like a rough lenear dependency of D ~ E. The exponent stays close to 1 meaning diffusion is a good description of the motion.
