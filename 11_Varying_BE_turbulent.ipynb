# Varying B and E in turbulent field

We now will vary B and E together to see the combined effect. 

```python
%matplotlib inline

from Diffulyzer import *

num_candidates = 300
num_samples = 200
max_trajectory_length = 100 * Mpc
grid_size = 256
grid_spacing = 30 * kpc
lMin = 100 * kpc
lMax = 1000 * kpc
alpha = -11./3.

Brms_min = 10 * nG
Brms_max = 100 * nG
E_min = 1 * EeV
E_max = 10 * EeV
variation_B_steps = 5
variation_E_steps = 10
variation_repeat = 3
```

With the above settings we run a bunch of simulations. We will have an outer loop varying the B and and the E in an inner loop, being in the [lMin, lMax] interval.

```python
delta_B = (Brms_max - Brms_min) / (variation_B_steps-1)
delta_E = (E_max - E_min) / (variation_E_steps-1)
series = []
for i in range(variation_B_steps):
    Brms = Brms_min + i * delta_B
    Es = []
    r_L = []
    Ds = []
    exponents = []
    for j in range(variation_E_steps):
        E = E_min + j * delta_E
        for k in range(variation_repeat):
            nr = 'Fit ' + str((i*variation_E_steps*variation_repeat)+(j*variation_repeat)+k+1) + '/' + str(variation_B_steps*variation_E_steps*variation_repeat)
            plot_title = nr + ', Brms=' + str(Brms / nG) + ' nG, E=' + str(E / EeV) + ' EeV'
            Es.append(E / EeV)
            r_L.append(larmor_radius(E, Brms) / kpc)
            data = sim(num_candidates, num_samples, max_trajectory_length, E, E, grid_size, grid_spacing, Brms, lMin, lMax, alpha)
            D, exponent = plot_rsq(data, plot_title)
            Ds.append(D)
            exponents.append(exponent)
    series.append((Brms / nG, Es, r_L, Ds, exponents))
```

An overview of all results shows the dependency on B and E.

```python
figure(figsize=(9, 5))
title('D ~ Brms, E')
labels = []
for B, Es, r_L, Ds, exponents in series:
    scatter(r_L, Ds, s=50)
    labels.append('B = ' + str(B) + ' nG')
grid()
ylabel('D')
xlabel('r_L [kpc]')
legend(labels)
show()

figure(figsize=(9, 5))
title('exponent ~ Brms, E')
for B, Es, r_L, Ds, exponents in series:
    scatter(r_L, exponents, s=50)
grid()
ylabel('exponent')
xlabel('r_L [kpc]')
legend(labels)
show()      
```
